# OOP

## Klases atribūti un metodes

```python
class Student:
    def __init__(self, name, course):
        self.name = name
        self.course = course
Student1 = Student("Jane", "JavaScript")
Student2 = Student("John", "Python")
print(Student1.name) # Jane
print(Student2.name) # John
```

## Klases veidošana

```python
class Dog:
# atribūti
    def __init__(self, name, age):
        self.name = name
        self.age = age
# metodes
    def sit(self):
        print(f"{self.name} is now sitting.")
    def roll_over(self):
        print(f"{self.name} rolled over!")
```

## Instances veidošana un metožu izsaukšana

```python
# Instances veidošana
my_dog = Dog('Willie', 6)

# Pieklušana atribūtiem ar punkta notāciju
print(f"My dog's name is {my_dog.name}.")
print(f"My dog is {my_dog.age} years old.")

# Metožu izsaukšana
my_dog.sit()
my_dog.roll_over()
```

## Vairākas instances

```python
my_dog = Dog('Willie', 6) # Prīma instance
your_dog = Dog('Lucy', 3) # Otra instance

print(f"My dog's name is {my_dog.name}.")
print(f"My dog is {my_dog.age} years old.")
my_dog.sit()
print(f"\nYour dog's name is {your_dog.name}.")
print(f"Your dog is {your_dog.age} years old.")
your_dog.sit()
```
