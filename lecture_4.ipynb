# OOP

## Klases atribūti un metodes

```python
class Student: #klase sākas ar lielo burtu, atšķirībā no fukcijas nav iekavas, tā vie'ta ir objekts def (nav fuknkcja, bet konstruktors)
    def __init__(self, name, course): #konstruktorS. Objekts init inicializē klasi. Iekavās self nozīmē, ka attiecās uz klasi Student
        self.name = name # paremetrus ar self piešķir atribūtus. Ar selft nav jānorāda visi paremetri, pietiek, ka tikai self norāda
        self.course = course #piekļūšana elementiem caur punktu (.)
Student1 = Student("Jane", "JavaScript") #tiek saglabāts mainīgais
Student2 = Student("John", "Python")
print(Student1.name) # Jane . Kad izveidota instanse Student1, tad caur punktu izsaucot atribūtu name, tiek iedots vārds. Labi noder DB, ka reģistrācija notiek, ierkastiem nav vajadzīgs for cikls
print(Student2.name) # John
```

## Klases veidošana

```python
class Dog: #svarīgi ievērto atkāpes
# atribūti
    def __init__(self, name, age):
        self.name = name #piešķirti parametri
        self.age = age
# metodes
    def sit(self): #fukcija sēdēt
        print(f"{self.name} is now sitting.")
    def roll_over(self): #fukcija kustēties
        print(f"{self.name} rolled over!")
```

## Instances veidošana un metožu izsaukšana

```python
# Instances veidošana
my_dog = Dog('Willie', 6) #paņemts no klases Dog vārdu + iedoti argumenti, kuri tiek pieškirti parametriem

# Pieklušana atribūtiem ar punkta notāciju
print(f"My dog's name is {my_dog.name}.") #no instances ar name iegūst aribūtu
print(f"My dog is {my_dog.age} years old.")

# Metožu izsaukšana
my_dog.sit()
my_dog.roll_over()
```

## Vairākas instances

```python
my_dog = Dog('Willie', 6) # Pirmā instance
your_dog = Dog('Lucy', 3) # Otrā instance

print(f"My dog's name is {my_dog.name}.")
print(f"My dog is {my_dog.age} years old.")
my_dog.sit()
print(f"\nYour dog's name is {your_dog.name}.")
print(f"Your dog is {your_dog.age} years old.")
your_dog.sit()
```
