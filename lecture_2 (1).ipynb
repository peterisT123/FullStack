#### Elementu pievienošana sarakstam
##### *append(elem) – elementa pievienošana beigās*

```python
motorcycles.append('ducati')
print(motorcycles)

motorcycles = []
motorcycles.append('honda')
motorcycles.append('yamaha')
motorcycles.append('suzuki')
print(motorcycles)
```

##### *insert(pos, elem) – elementa pievienošana pēc pozīcijas*

```python
motorcycles = ['honda', 'yamaha', 'suzuki']
motorcycles.insert(0, 'ducati')
print(motorcycles)
```

#### Elementu izdzēšana no saraksta
##### *elem = pop() – elementa izmešana no beigām (ar tā atgriešanu)*

```python
motorcycles = ['honda', 'yamaha', 'suzuki']
print(motorcycles)
popped_motorcycle = motorcycles.pop()
print(motorcycles)
print(popped_motorcycle)

#Funkcija del darbojas līdzīgi kā piešķiršana – var izdzēst vienu elementu vai intervālu:

motorcycles = ['honda', 'yamaha', 'suzuki', 'KTM']
del motorcycles[0:2]
print(motorcycles)
```

#### Elementu izdzēšana no saraksta
##### *elem = pop(pos) – elementa izmešana pēc pozīcijas (var būt negatīva)*

```python
first_owned = motorcycles.pop(0)
print(f"The first motorcycle I owned was a {first_owned.title()}.")
# remove(val) – elementa (pirmā sastaptā) izmešana pēc vērtības

motorcycles = ['honda', 'yamaha', 'suzuki', 'ducati']
print(motorcycles)
motorcycles.remove('ducati')
print(motorcycles)
motorcycles = ['honda', 'yamaha', 'ducati', 'suzuki', 'ducati']
print(motorcycles)
motorcycles.remove('ducati')
print(motorcycles)
```

#### Vārdnīcas

```python
alien_0 = {'color': 'green', 'points': 5}
print(alien_0['color'])
print(alien_0['points'])
```

#### Vārdnīcas izveidošana

```python
#Tukšu vārdnīcu izveido vai nu, izmantojot funkciju dict, vai ar tukšām figūriekavām {}.
dictionary_1 = {}
disctionary_2 = dict()
print(dictionary_1, len(dictionary_1), type(dictionary_1))
print(disctionary_2, len(disctionary_2), type(disctionary_2))
#Ar komatiem atdalītu pāru (atslēga:vērtība) virkne figūriekavās nozīmē vārdnīcas izveidošanu.
dd = {'četri': 4 , 'divi': 2, 'trīs':3, 'viens':1}
print(dd)
print(len(dd))
print(list(dd.keys()))
```

#### Piekļūšana vārdnīcas ierakstiem

```python
# Piekļūšana vārdnīcas ierakstiem notiek pa vienam, izmantojot atslēgu ar tādu pašu sintaksi [ ], kā ar indeksu piekļūstot saraksta elementam.
alien_0 = {'color': 'green'}
print(alien_0['color'])
alien_0 = {'color': 'green', 'points': 5}
new_points = alien_0['points']
print(f"You just earned {new_points} points!")
#Piekļūt vārdnīcas elementiem var ari izmantojot metodi get()
alien_0 = {'color': 'green', 'speed': 'slow'}
point_value = alien_0.get('color')
print(point_value)
```

#### Ierakstu pievienošana vārdnīcai

```python
#Viena ieraksta pievienošana vārdnīcai tiek veikta ar operatoru [ ]:

dd = {}
dd[3]='three'
dd[5]='five'
print(dd)

alien_0 = {'color': 'green', 'points': 5}
alien_0['x_position'] = 0
alien_0['y_position'] = 25
print(alien_0)
```

#### Ierakstu izdzēšana no vārdnīcas

```python
# Ieraksta izdzēšanai no vārdnīcas lieto operatoru del vai metodi pop.

dd = {'četri':(4,'four'), 'divi':(2,'two'),
'trīs':(3,'three'), 'viens':(1,'one')}
del dd['divi']
dd.pop('trīs')
print(dd)
```

#### Vārdnīcas vērtību izmaiņa 

```python
# Lai izmainītu vārdnīcas vērtību, 
# jānorada vārdnīcas vārdu ar atslēgu kvadrātiekavas,
# un pēc tam jauno vērtību kuru velties piešķirt tai atslēgai. 

alien_0 = {'color': 'green'}
print(f"The alien is {alien_0['color']}.")
alien_0['color'] = 'yellow'
print(f"The alien is now {alien_0['color']}.")
```
