## Neobligātie argumenti

```python
def get_formatted_name(first_name, last_name, middle_name=''): # middle_name var pieškirt tukšu vikrnī, lai ir true IF definājot
    if middle_name:
        full_name = f"{first_name} {middle_name} {last_name}"
    else:
        full_name = f"{first_name} {last_name}"
    return full_name.title()

musician = get_formatted_name('jimi', 'hendrix',  "james")
print(musician)
```

## *args, **kwargs

```python
def make_pizza(*args):   #args priekš kad nezinām cik paramteri būs, te izveidojoas tukšs slēgtais saraksts ()
    print(args)

make_pizza('pepperoni')
make_pizza('mushrooms', 'green peppers', 'extra cheese')
```

```python
def print_kwargs(**kwargs): #atslēgas vārda argumenti, tiek veidota vārdnīca ne saraksts ar atslēgu un vērtību
        print(kwargs)

print_kwargs(kwargs_1="Shark", kwargs_2=4.5, kwargs_3=True)

def print_values(**kwargs):
    for key, value in kwargs.items():
        print(f"The value of {key} is {value}")

print_values(my_name="Sammy", your_name="Casey")
```

## funkcijas ar if-else un for

```python
def my_function(x, y, z=1.5):
    if z > 1:
        return z * (x + y)
    else:
        return z / (x + y)

a = my_function(5, 6, z=0.7)
print(round(a, 2))
b= my_function(3.14, 7, 3.5)
print(b)
c = my_function(10, 20)
print(c)
```

## lambda

```python
# lambda arguments : expression  #funkcija bez nosaukuma, ļauj taupīt kodu

x = lambda a : a + 10
print(x(5))
```

```python
x = lambda a, b : a * b
print(x(5, 6))
```

```python
x = lambda a, b, c : a + b + c
print(x(5, 6, 2))
```

```python
def myfunc(n):
  return lambda a : a * n

mydoubler = myfunc(2)

print(mydoubler(11))
```

```python
#ar map() funkciju vienlaicīgi nekā for ciklam
nums = [1,2]
ssss = map(lambda x: x*2, nums)
print(list(ssss))
sssquere = map(lambda x: x**2, nums)
print(list(sssquere))
```

```python
def myfunc(n):
  return lambda a : a * n

mydoubler = myfunc(2)
mytripler = myfunc(3)

print(mydoubler(11))
print(mytripler(11))
```

```python
def apply_to_list(some_list, f):
    return [f(x) for x in some_list]

ints = [4, 0, 1, 5, 6]
apply_to_list(ints, lambda x: x * 2)
```
